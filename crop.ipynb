### Encoding

```python
crop['label'].unique()
```

```python
crop_dict={
    'rice':1,
    'maize':2,
    'chickpea':3,
    'kidneybeans':4,
    'pigeonpeas':5,
    'mothbeans':6,
    'mungbean':7,
    'blackgram':8,
    'lentil':9,
    'pomegranate':10,
    'banana':11,
    'mango':12,
    'grapes':13,
    'watermelon':14,
    'muskmelon':15,
    'apple':16,
    'orange':17,
    'papaya':18,
    'coconut':19,
    'cotton':20,
    'jute':21,
    'coffee':22
}
```

```python
crop['crop_no']=crop['label'].map(crop_dict)
crop.head()
```

```python
crop.tail()
```

```python
crop.drop('label', axis=1, inplace=True)
crop.tail() 
```

```python
# Train Test Split
x=crop.drop('crop_no', axis=1)
y=crop['crop_no']
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train) 
x_test_scaled=scaler.transform(x_test)
```

```python
# Training the model
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train_scaled, y_train)
```

```python
y_pred=dtc.predict(x_test_scaled)
acs=accuracy_score(y_test, y_pred)
print('Accuracy Score:', acs)
```

```python
#Prediction
def crop_recommendation(n, p, k, temp, humidity, ph, rainfall):
    n_scaled=scaler.transform([[n, p, k, temp, humidity, ph, rainfall]])
    prediction=dtc.predict(n_scaled).reshape(1,-1)
    crop_dict={
        1:'rice',
        2:'maize',
        3:'chickpea',
        4:'kidneybeans',
        5:'pigeonpeas',
        6:'mothbeans',
        7:'mungbean',
        8:'blackgram',
        9:'lentil',
        10:'pomegranate',
        11:'banana',
        12:'mango',
        13:'grapes',
        14:'watermelon',
        15:'muskmelon',
        16:'apple',
        17:'orange',
        18:'papaya',
        19:'coconut',
        20:'cotton',
        21:'jute',
        22:'coffee'
    }
    crop=[crop_dict[i] for i in prediction[0]]
    return f"{crop} is a suitable crop for the given conditions"
    
        
```

```python
#predicting the crop
crop_recommendation(90, 32, 41, 10.879744, 92.002744, 4.502985, 201.935536)
```
